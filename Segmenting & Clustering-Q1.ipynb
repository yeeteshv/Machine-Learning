Step1 Scrape wikipedia page

```python
import pandas as pd
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

Our desired table is found at index 0

```python
step2=df[0]
```

Ignore records where Borough is 'Not assigned'

```python
step2=df[0][df[0].Borough !='Not assigned']
```

Replace Neighbourhood with Borough for records where Neighhourhood is 'Not assigned'

```python
step3=step2
step3.Neighbourhood = step3.Borough.where(step3.Neighbourhood == 'Not assigned', step2.Neighbourhood)
step3
```

Define a custom grouping method for string type

```python
foo = lambda val: ", ".join(val)
```

Apply grouping on the basis of Postcode and Borough

```python
step3=step3.groupby(by=['Postcode','Borough']).agg({'Neighbourhood': foo}).reset_index()
```
